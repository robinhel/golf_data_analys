## Bakgrund och syfte

Datasetet kommer från Kaggle och innehåller statistik från PGA Tour säsongerna 2010–2018. Varje rad representerar en spelares samlade statistik för varje år.

Jag är intresserad av golf och använde det här datasetet för att försöka svara på en fråga som faktiskt berör mitt eget spel vad är det egentligen som avgör om man presterar bra? Slår man för lite med drivern, lägger för mycket tid på putting-greenen, eller är det järnspelet som är nyckeln?

Populationen är professionella spelare på PGA Tour världens högsta nivå. Det begränsar hur direkt jag kan översätta insikterna till mitt eget spel, men mönstren kan ändå peka i en riktning.

**Kolumner av intresse:**
- `Avg Distance` – genomsnittlig driving distance i yards
- `Fairway Percentage` – andel drives som landar i fairway
- `Average Score` – genomsnittlig score per runda
- `Wins` – antal vinster under säsongen
- `SG:OTT` – Strokes Gained Off The Tee (driving)
- `SG:APR` – Strokes Gained Approach (järnspel)
- `SG:ARG` – Strokes Gained Around The Green (nearspel)
- `Average SG Putts` – Strokes Gained Putting

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pgaTourData.csv')


df.head()
df.info()
df.describe()
```

## Datatvätt

**Rounds (634 saknade rader):** Spelare utan runddata saknar all spelstatistik, ta bort.

**Money och Points:** Lagrade som text 2,680,487. Rensar bort och kommatecken och konverterar till tal. Används inte i visualiseringarna så kvarstående NaN spelar ingen roll.

**Wins (2 019 NaN) och Top 10 (854 NaN):** Fylls i med 0. Rimligt eftersom de flesta spelare inte vinner eller toppar under en säsong men det är ett antagande.

Inga dubletter. Inga orimliga värden.

## Datatvätt

```python
# 1. Ta bort rader som saknar all spelstatistik (ingen Rounds = ingen speldata)
df = df.dropna(subset=['Rounds'])

# 2. Konvertera Money från text till tal
df['Money'] = df['Money'].str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df['Money'] = pd.to_numeric(df['Money'], errors='coerce')

# 3. Konvertera Points från text till tal
df['Points'] = df['Points'].str.replace(',', '', regex=False)
df['Points'] = pd.to_numeric(df['Points'], errors='coerce')

# 4. Fyll i 0 för Wins och Top 10
# Antagande: NaN = 0 vinster/placeringar, inte saknad data.
# Rimligt eftersom de flesta spelare inte vinner under en säsong.
df['Wins'] = df['Wins'].fillna(0)
df['Top 10'] = df['Top 10'].fillna(0)



df.head()

df.info()

df.describe()


```

# kan korta slag vara fördelsaktiga?

```python
scatter_df = df[["Player Name", "Avg Distance", "Average Score"]].dropna()

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(scatter_df["Avg Distance"], scatter_df["Average Score"], alpha=0.3, color="#2e86c1")

tiger = scatter_df[scatter_df["Player Name"] == "Tiger Woods"].iloc[0]
ax.annotate(
    "Tiger Woods",
    xy=(tiger["Avg Distance"], tiger["Average Score"]),
    xytext=(tiger["Avg Distance"] + 5, tiger["Average Score"] - 0.5),
    arrowprops=dict(arrowstyle="->", color="red"),
    color="red"
)

ax.set_title("Längre drive leder inte alltid till bättre score")
ax.set_xlabel("Driving distance (yards)")
ax.set_ylabel("Genomsnittlig score (slag)")
```

# kortare slag = mer fairways?

```python
fairway_df = df[["Player Name", "Avg Distance", "Fairway Percentage"]].dropna()

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(fairway_df["Avg Distance"], fairway_df["Fairway Percentage"], alpha=0.3, color="#2e86c1")

tiger = fairway_df[fairway_df["Player Name"] == "Tiger Woods"].iloc[0]
ax.annotate(
    "Tiger Woods",
    xy=(tiger["Avg Distance"], tiger["Fairway Percentage"]),
    xytext=(tiger["Avg Distance"] + 5, tiger["Fairway Percentage"] + 2),
    arrowprops=dict(arrowstyle="->", color="red"),
    color="red"
)

ax.set_title("Längre drive = sämre fairway träff")
ax.set_xlabel("Driving distance (yards)")
ax.set_ylabel("Fairway träff (%)")
```

## Avslutning

Järnspel och driving korrelerar mer med vinster än putting "putt for dough" verkar inte stämma på den här nivån. Längre drives ger sämre fairway träff, och toppspelarna skiljs åt med bråkdelar av ett slag.
